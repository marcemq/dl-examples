# Plotting make_moons dataset

```
# Package imports
import matplotlib.pyplot as plt
import numpy as np
import sklearn.datasets

# Generate a make_moon dataset and plot it
np.random.seed(22)
X, y = sklearn.datasets.make_moons(300, noise=0.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.RdBu)

# Add title and axis names
plt.title('make_moons data')
plt.xlabel('x')
plt.ylabel('y')
```
